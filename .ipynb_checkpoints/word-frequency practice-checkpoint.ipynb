```python
def remove_punctuation_and_lines(x):
    x = x.lower()
    for c in string.punctuation:
        x = x.replace(c, "")
    for c in string.whitespace:
        x = x.replace(c, " ")
    x = x.split()
    return x
```

```python
remove_punctuation_and_lines(book)
```

```python
def word_dictionary(x):
    word_dict = {}
    for word in x:
        if word in word_dict.keys():
            word_dict[word] += 1
        else:
            word_dict[word] = 1
    return word_dict
```

```python
y = remove_punctuation_and_lines(book)
word_list(y)
```

```python
y = remove_punctuation_and_lines(book)
z = word_dictionary(y)
word_list = []
for key, value in z.items():
    word = [key, value]
    word_list.append(word)
    #print(word)
    
sorted_list = sorted(word_list, reverse = True, key= lambda x: x[1] )[:20]
for word, appears in sorted_list:
    print("{}-- {}".format(word, appears))
#print(sorted_list)
```

```python
def word_list(dictionary):
    word_list = list(dictionary.items())
    return word_list
    
```

```python
just_letters = remove_punctuation_and_lines(book)
word_dict = word_dictionary(just_letters)
unsorted_list = word_list(word_dict)
print(unsorted_list)
```

```python
sorted_list = sorted(word_list, reverse = True, key= lambda x: x[1] )[:20]
for word, appears in sorted_list:
    print("{}-- {}".format(word, appears))
```

```python
def sort_list(a_list, how_many):
    sorted_list = sorted(a_list, reverse = True, key= lambda x: x[1])[:how_many]
    for word, appears in sorted_list:
        print("{} {}".format(word, appears))
```

```python
just_letters = remove_punctuation_and_lines(book)
word_dict = word_dictionary(just_letters)
unsorted_list = word_list(word_dict)
final_list = sort_list(unsorted_list, 5)
```
